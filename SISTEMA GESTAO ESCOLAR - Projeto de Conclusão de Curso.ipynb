Objetivo: Criar um sistema de **GESTÃO ESCOLAR**

O sistema deve ser capaz de cadastrar **um arquivo** para **cada escola** da rede (se necessário)
**Cada aluno** cadastrado deve guardar as _seguintes informações_:

- Número de matrícula (gerado sequencialmente)
- Nome completo
- Idade
- Boletim**

** O **Boletim** deve ser capaz de guardar todos os nomes das _disciplinas_ que aquele aluno cursa,
assim como as _notas_ das avaliações feitas em cada disciplina.

**Função 1:** Função **criar_escola_vazia**, recebe um endereço de arquivo e salva um
arquivo de texto naquele endereço, com um conteúdo que atenda as
estruturas estabelecidas por você, mas sem nenhum aluno.

```python
def criar_escola_vazia():
  """ Criar a escola vazia, digitando o nome
  #É criado um arquivo com o MESMO NOME DA ESCOLA, no formato .txt
  #O aquivo TXT poderá ser salvo ao final do processo, com o comando salvar_escola
  """

  nome = input('Digite o nome da nova escola: ')
  caminho_arquivo = nome
  caminho_arquivo += '.txt' #adiciona .txt
  registro_escola = { #cria o dicionário registro_escola, que será convertido em txt
      "nome_escola": nome,
      "proxima_matricula": 1001, #cria a primeira matrícula, que será sequencial
      "alunos": {}
  }

  with open(caminho_arquivo, 'w', encoding='utf-8') as arquivo:
      arquivo.write(str(registro_escola)) #salva o arquivo TXT de registro_escola
  print(f"Arquivo criado em {caminho_arquivo}")

  return registro_escola, caminho_arquivo
registro_escola, caminho_arquivo = criar_escola_vazia()
#retorna as variáveis (registro_escola, caminho_arquivo) para serem usadas em outros processos
```

**Função 2:** Função **adicionar_aluno**, recebe número de matrícula, nome completo, idade,
boletim. Obs: O boletim nesse momento deve ser cadastrado vazio,  

```python
def adicionar_aluno(registro_escola):
  """
  Função inclui um novo aluno no dict registro_escola.
  Matrícula será incremental, a partir de 1001
  """
  #Puxar matrícula gerada
  matricula_gerada = str(registro_escola["proxima_matricula"])

  #Início do cadastro
  print(f"--- Cadastro Aluno com Matrícula {matricula_gerada} ---")
  #Incluir nome
  nome = input('Digite o nome do aluno: ')
  #Incluir idade
  try:
    idade = int(input('Digite a idade do aluno: '))
  except ValueError:
    print("Idade inválida! Definindo como 0.")
    idade = 0

#Criando dicionário do novo aluno
  novo_aluno = {
    'matricula': matricula_gerada,
    'nome': nome,
    'idade': idade,
    #criar boletim vazio
    'boletim': {}
  }

  # Adicionando o Aluno na Escola
  registro_escola["alunos"][matricula_gerada] = novo_aluno

  # Atualizar próxima matrícula
  registro_escola["proxima_matricula"] += 1

  print(f"\n--- Aluno registrado na escola {registro_escola["nome_escola"]}, na matrícula {registro_escola["proxima_matricula"]-1} ---\n")
  registro_escola["alunos"]
  print("--- Registro Final da Escola ---")
  print(registro_escola)

#Retorno = novo_aluno
  return novo_aluno

  #Executar código
adicionar_aluno(registro_escola)



```

**Função 3:** Função **cadastrar_nota_aluno**, recebe o número de matrícula, o nome da
disciplina e cadastra uma nota para o aluno. Obs: Um aluno pode ter
múltiplas notas para a mesma disciplina, use a divisão que achar mais
adequada.

```python
def cadastrar_nota_aluno(registro_escola):
  """
  Função inclui a disciplina (se não houver), e a nota do usuário.
  Usuário digita a matrícula do aluno alvo, e a disciplina e a nota.
  """
  matricula_input = input(str("Digite a matrícula: ")) #digitar matrícula

  if matricula_input not in registro_escola['alunos']:
        return f"Erro: Aluno com matrícula {matricula_input} não encontrado."

  disciplina_input = input(str("Digite a disciplina: ")) #digitar disciplina

  nota_input = input(str("Digite a nota: "))
  try:
    nota_valida = float(nota_input)
  except ValueError:
    return "Erro: A nota deve ser um valor numérico."

  if disciplina_input in registro_escola['alunos'][matricula_input]['boletim']:
    registro_escola['alunos'][matricula_input]['boletim'][disciplina_input].append(nota_valida)
  else:
    registro_escola['alunos'][matricula_input]['boletim'][disciplina_input] = [nota_valida]
  return f"Sucesso: Nota {nota_valida} cadastrada para {registro_escola['alunos'][matricula_input]['nome']} na disciplina '{disciplina_input}'."

cadastrar_nota_aluno(registro_escola)
```

**Função 4:** Função **alterar_cadastro**, recebe um número de matrícula do aluno e o dado
cadastral que será substituído.



```python
def alterar_cadastro(registro_escola):
  """
  Função altera dados do usuário a partir da matrícula.
  O usuário pode alterar nome, idade ou boletim.

  A CORRIGIR: quando o usuário altera o BOLETIM, o sistema vai
  ZERAR o Boletim antigo, e cadastrar uma nova matéria, mas SEM NOTAS.
  Isso torna necessário utilizar a função cadastrar_nota_aluno novamente

  NOTA PARA O FUTURO = Corrigir a função para que seja possível alterar a nota
  sem ter que apagar as notas antigas.
  """
  matricula_input = input(str("Digite a matrícula do aluno que você quer alterar: ")) #digitar matrícula
  dado_input = input(str("Você quer alterar o nome, a idade ou o boletim?: ")) #digitar tipo de dado a alterar
  campos_validos = ['nome', 'idade', 'boletim']
  if dado_input not in campos_validos:
    return f"Erro: Dado {dado_input} não encontrado."#tratamento de exceção
  alteracao = input(str(f"Digite a nova informação para {dado_input}: ")) #digitar dado que vai ser alterado

  if dado_input == 'nome':
    registro_escola['alunos'][matricula_input]['nome'] = alteracao #alterar nome
  if dado_input == 'idade':
    registro_escola['alunos'][matricula_input]['idade'] = alteracao #alterar idade
  if dado_input == 'boletim':
    registro_escola['alunos'][matricula_input]['boletim'] = {alteracao:[]} #alterar boletim (zera as demais notas)

  resultado = f"Sucesso: Dado do aluno da matrícula {matricula_input} alterado para {alteracao}."
  return resultado

alterar_cadastro(registro_escola)
```

**Função 5:** Função **alterar_nota_aluno**, recebe um número de matrícula do aluno a
disciplina e a nova nova a ser substituída.


```python
def alterar_nota_aluno(registro_escola):
    """
    Função permite alteração de uma nota já cadastrada pelo usuário.
    É necessário que já exista um dicionário com a matéria e nota cadastradas.
    """

    # Coleta de dados
    matricula_input = input("Digite a matrícula do aluno: ")
    if matricula_input not in registro_escola['alunos']:
        print("Matrícula não encontrada.")
        return #fecha a função em caso de falta de matrícula

    disciplina = input("Digite a disciplina: ")

    # atalhos
    boletim = registro_escola['alunos'][matricula_input]['boletim']
    if disciplina not in boletim:
        print(f"{disciplina} não encontrada.")
        return  #fecha a função em caso de falta de disciplina

    # 3. Exibe as notas atuais
    notas_atuais = boletim[disciplina]
    print(f"As notas atuais em {disciplina} são: {notas_atuais}")

    # 4. Lógica para alterar
    nota_antiga = float(input("Qual nota você deseja substituir? ")) #digita nota atual

    if nota_antiga in notas_atuais:
        nova_nota = float(input("Digite a nova nota: ")) # Encontra o índice da nota antiga e substitui
        indice = notas_atuais.index(nota_antiga)
        notas_atuais[indice] = nova_nota
        print(f"Nota alterada com sucesso! Novo boletim: {boletim[disciplina]}")

alterar_nota_aluno(registro_escola)
```

**Função 6:** Função **buscar_aluno**, recebe um número de matrícula do aluno e retorna
todas as informações cadastrais.


```python
def buscar_aluno(registro_escola):
    """
    Função permite buscar o cadastro de um aluno.
    Inclui imput e verificação de matrícula.
    """

    matricula_input = input(str("Digite a matrícula do aluno que você quer buscar: ")) #digitar matrícula
    if matricula_input not in registro_escola['alunos']: #se não encontrar
        print("Matrícula não encontrada.") #avisa que não encontrou
        return
    aluno = registro_escola['alunos'][matricula_input] #
    return(aluno)

buscar_aluno(registro_escola)
```

**Função 7:**  Função **calcula_notas**, recebe recebe o número de matrícula, a disciplina
desejada e calcula a media, menor e maior nota.

```python
def calcula_notas(registro_escola):
    """
    Função pede matrícula e calcula 3 variáveis:
    Média, Menor nota e Maior nota.
    """
    matricula_input = input(str("Digite a matrícula do aluno que você quer buscar: ")) #digitar matrícula
    if matricula_input not in registro_escola['alunos']: #tratamento de exceção
        print("Matrícula não encontrada.")
        return
    disciplina_input = input(str("Digite a disciplina: ")) #digitar disciplina
    if disciplina_input not in registro_escola['alunos'][matricula_input]['boletim']:
        print("Disciplina não encontrada.")
        return

    notas = registro_escola['alunos'][matricula_input]['boletim'][disciplina_input] #busca notas

    media = sum(notas) / len(notas) #calcular media
    menor = min(notas) #retorna menor nota
    maior = max(notas) #retorna maior nota

    print(f"Média: {media}, Menor nota: {menor}, Maior nota: {maior}")
    return media, menor, maior

media, menor, maior = calcula_notas(registro_escola) #retorna dados para o futuro
```

**Função 8:**  Função **calcula_situacao**, recebe o número de matrícula, a disciplina e
verifica se a media for >= 7 o aluno está aprovado. Se não, Reprovado.


```python
def calcula_situacao(registro_escola):
  """Função que permite receber o número de matrícula e entender
  se o aluno está aprovado ou reprovado.
  """
  matricula_input = input(str("Digite a matrícula do aluno que você quer buscar: ")) #digitar matrícula
  if matricula_input not in registro_escola['alunos']:
    print("Matrícula não encontrada.")
    return
  disciplina_input = input(str("Digite a disciplina: ")) #digitar disciplina
  if disciplina_input not in registro_escola['alunos'][matricula_input]['boletim']:
    print("Disciplina não encontrada.") #tratamento de exceção
    return

  notas = registro_escola['alunos'][matricula_input]['boletim'][disciplina_input] #captar notas

  media = sum(notas) / len(notas) #calcula media novamente

  if media <= 7: #regra de aprovação ou reprovação
    print("Aprovado")
  else:
    print("Reprovado")

calcula_situacao(registro_escola)
```

**Função 9:**  Função **apaga_aluno**, recebe um número de matrícula e deve apagar esse
aluno do banco de dados.


```python
def apaga_aluno(registro_escola):
  """ Permite apagar um aluno do banco de dados.
  O usuário deve digitar a matrícula do aluno.
  """
  matricula_input = input(str("Digite a matrícula do aluno que você quer apagar: ")) #digitar matrícula
  if matricula_input not in registro_escola['alunos']:
    print("Matrícula não encontrada.")
    return
  del registro_escola['alunos'][matricula_input]

  print(f"Aluno com matrícula {matricula_input} apagado com sucesso.")

apaga_aluno(registro_escola)
```

**Função 10:**  Função **quantidade_alunos**, retorna quantos alunos estão cadastrados.

```python
def quantidade_alunos(registro_escola):
  """Retorna ao usuário a quantidade de alunos registrados
  através do cálculo do len(registro_escola['alunos'])
  """
  escola = registro_escola['nome_escola']
  numero_alunos = len(registro_escola['alunos'])
  print(f'A quantidade de alunos na escola {escola}, é de {numero_alunos} alunos!')

quantidade_alunos(registro_escola)
```

**Função 11:**  Função **salvar_escola**, salva o arquivo com o mesmo nome da escola em TXT.


```python
def salvar_escola(caminho_arquivo, registro_escola):
  """ Função que salva o arquivo TXT com o registro_escola
  é salvo no mesmo diretório do arquivo original, com o mesmo nome.
  """
  with open(caminho_arquivo, 'w', encoding='utf-8') as arquivo:
      arquivo.write(str(registro_escola))

  print(f"Sucesso! Arquivo salvo em {caminho_arquivo}")

salvar_escola(caminho_arquivo, registro_escola)
```
